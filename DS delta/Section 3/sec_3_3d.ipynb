# Python Principal Component Analysis (PCA) for 3D data

## 1. Plot the 3D data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D # Enable to display 3d data

arr2d_x = np.loadtxt("pca3d.csv", delimiter=",")

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(20, 33) # Set the viewing angle
ax.scatter3D(arr2d_x[:, 0], arr2d_x[:, 1], arr2d_x[:, 2], c=arr2d_x[:, 2]) # Plot 3d plot
```

## 2. PCA for 3D data

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(arr2d_x)

arr2d_score = pca.transform(arr2d_x)
print(pca.components_)
print(arr2d_score)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(0, 30)
ax.scatter3D(arr2d_x[:, 0], arr2d_x[:, 1], arr2d_x[:, 2], c=arr2d_x[:, 2])
ar_line0 = np.array([-4*pca.components_[0], 4*pca.components_[0]])
ar_line1 = np.array([-3*pca.components_[1], 3*pca.components_[1]])
ax.plot(ar_line0[:, 0], ar_line0[:, 1], ar_line0[:, 2], linewidth=1, c="b")
ax.plot(ar_line1[:, 0], ar_line1[:, 1], ar_line1[:, 2], linewidth=1, c="g")
```

## 3. Display contribution ratio and scatter plot of PC score

```python
print("Contribution ration (1st 2nd): ", pca.explained_variance_ratio_)

plt.figure(figsize=(6, 6))
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.scatter(arr2d_score[:, 0], arr2d_score[:, 1])
```
