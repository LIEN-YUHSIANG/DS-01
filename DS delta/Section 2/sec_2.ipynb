# Clustering with python

## 1. Generate data for clustering

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

mu0, mu1 = np.array([0, 5]), np.array([5, 0])
cov0 = cov1 = np.array([[4, 0], [0, 4]])
np.random.seed(0)
n0 = n1 = 100

arr2d_x0 = np.random.multivariate_normal(mu0, cov0, n0)
arr2d_x1 = np.random.multivariate_normal(mu1, cov1, n1)
arr2d_x = np.concatenate([arr2d_x0, arr2d_x1])
array_y = np.concatenate([np.zeros(n0), np.ones(n1)])

plt.figure(figsize=(6, 6))
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=array_y, cmap='bwr')

```

## 2. Clustering using the EM algorith with SK learn

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=2, random_state=0)
gmm.fit(arr2d_x)
arr2d_pp = gmm.predict_proba(arr2d_x)
plt.figure(figsize=(6, 6))
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=arr2d_pp[:,1], cmap='bwr')

print(gmm.means_)
print(gmm.covariances_)
```

## 3. Modify the data generation code

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

mu0, mu1 = np.array([0, 5]), np.array([5, 0])
cov0 = np.array([[4, 2], [2, 3]])
cov1 = np.array([[2, -2], [-2, 5]])
np.random.seed(0)
n0 = n1 = 100

arr2d_x0 = np.random.multivariate_normal(mu0, cov0, n0)
arr2d_x1 = np.random.multivariate_normal(mu1, cov1, n1)
arr2d_x = np.concatenate([arr2d_x0, arr2d_x1])
array_y = np.concatenate([np.zeros(n0), np.ones(n1)])

plt.figure(figsize=(6, 6))
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=array_y, cmap='bwr')
```

## 4. Reexecute clustering with EM algorithm

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=2, random_state=0)
gmm.fit(arr2d_x)
arr2d_pp = gmm.predict_proba(arr2d_x)
plt.figure(figsize=(6, 6))
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=arr2d_pp[:,1], cmap='bwr')

print(gmm.means_)
print(gmm.covariances_)
```

## 5. Clustering real data

### 5-1. Plot out the original data

```python
# Plot out the original data

arr2d_data = np.loadtxt(fname='irisdata.csv', delimiter=",")
array_y = arr2d_data[:, 0]
arr2d_x = arr2d_data[:, 1:]

plt.figure(figsize=(6, 6))
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=array_y, cmap='brg')
```

### 5-2. Clustering using KMeans

```python
# Clustering using KMeans

from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, random_state=0)
km.fit(arr2d_x)

plt.figure(figsize=(6, 6))
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=km.labels_, cmap='brg')
```

### 5-3. Clustering using EM algorithm

```python
# Clustering using EM algorithm

from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=3, random_state=0)
gmm.fit(arr2d_x)
array_yhat = gmm.predict(arr2d_x)

plt.figure(figsize=(6, 6))
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=array_yhat, cmap='brg')
```
