# Hypothesis Testing and Data Transformation

## Normal distribution

```python
%matplotlib inline
# Show all output at notebook
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np 
import matplotlib.pyplot as plt 
n = 100
np.random.seed(0)
array_x = np.random.randn(n)
plt.hist(array_x, bins=10)
```

## Plot Q-Q plot

```python
from scipy import stats 

plt.figure(figsize=(6, 6))
stats.probplot(array_x, dist="norm", plot=plt)
```

## Perform Shapiro-Wilk Normality test

```python
stats.shapiro(array_x)
```

## Test out a uniform distribution

```python
np.random.seed(0)
array_x = np.round(np.random.uniform(0, 100, n))
plt.hist(array_x, bins=10)
```

```python
plt.figure(figsize=(6, 6))
stats.probplot(array_x, dist="uniform", sparams=(0, 100), plot=plt)
```

## Perfrom Kolmogorov-Smirnov's test since Shapiro-Wilk only works for normal distribution

```python
stats.kstest(array_x, "uniform", args=(0, 100))
```

# Data transformation

```python
array_x = np.loadtxt('score.csv', delimiter=",")
plt.hist(array_x, range=(0, 100), bins=10)
```

```python
array_xs = stats.zscore(array_x)
stats.shapiro(array_xs)

## Since tehe p-value is so small, we can conclude that the data do not follow normal distribution.
```

```python
array_z = np.log(100.01 - array_x)
array_zs = stats.zscore(array_z)
plt.hist(array_zs, bins=10)
```

```python
plt.figure(figsize=(6, 6))
stats.probplot(array_zs, dist="norm", plot=plt)
```

```python
stats.shapiro(array_zs)

# After data transformation, we can see that the p-value of shapiro test is high.
# Thus right now the transformed data is suit to normal distribution
```
