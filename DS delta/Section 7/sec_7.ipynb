```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt 

arr2d_data = np.loadtxt('irisdata.csv', delimiter=",")

array_y= arr2d_data[:, 0]
arr2d_x = arr2d_data[:, 1:]

plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=array_y, cmap="brg")
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=3)
dtc.fit(arr2d_x, array_y)
print(dtc.score(arr2d_x, array_y))

array_yhat = dtc.predict(arr2d_x)
print(array_y)
print(array_yhat)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(10, 10))
plot_tree(dtc, filled=True)
```

```python
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=array_y, cmap='brg')
x0, x1 = np.meshgrid(np.arange(0.5, 7.5, 0.02), np.arange(0, 3, 0.02))
y = dtc.predict(np.stack([x0.ravel(), x1.ravel()], axis=1))
yy = y.reshape(x0.shape)
plt.pcolormesh(x0, x1, yy, alpha=0.05, cmap='brg', shading="gouraud")
```
