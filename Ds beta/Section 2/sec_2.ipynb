```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

mu_alpha, mu_beta = 0, 0
sigma_alpha, sigma_beta = 2, 2

array_mean = np.array([mu_alpha, mu_beta])
arr2d_cov = np.array([[sigma_alpha**2, 0], [0, sigma_beta**2]])

array_alpha = np.arange(-6, 6, 0.01)
array_beta  = np.arange(-6, 6, 0.01)
arr2d_Alpha, arr2d_Beta = np.meshgrid(array_alpha, array_beta)

array_ab = np.c_[np.ravel(arr2d_Alpha), np.ravel(arr2d_Beta)]
arr2d_pdf = multivariate_normal.pdf(x=array_ab, mean=array_mean, cov=arr2d_cov)
arr2d_pdf = arr2d_pdf.reshape(arr2d_Alpha.shape)

plt.pcolormesh(arr2d_Alpha, arr2d_Beta, arr2d_pdf, cmap='GnBu')
plt.colorbar()

```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(7, 7), dpi=120)
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(arr2d_Alpha, arr2d_Beta, arr2d_pdf, cmap='GnBu')
fig.colorbar(surf)
```

```python
np.random.seed(0)
sigma = 1
n, alpha, beta = 3, 2, 1
array_x = np.array([1, -1, 0.7])
array_y = alpha * array_x + beta + np.random.normal(0, sigma, n)
print(array_y)
plt.scatter(array_x, array_y)
```

```python
sum_x = sum(array_x)
sum_y = sum(array_y)
sum_x2 = sum(array_x**2)
sum_xy = sum(array_x*array_y)
a = 1/(sigma_alpha**2)+sum_x2/(sigma**2)
d = 1/(sigma_beta**2)+n/(sigma**2)
b = sum_x/(sigma**2)
arr2d_cov = np.array([[d, -b], [-b, a]]/(a*d-(b**2)) )

f = mu_alpha/(sigma_alpha**2) + sum_xy/(sigma**2)
g = mu_beta/(sigma_beta**2) + sum_y/(sigma**2)

mu_nalpha = (d*f-b*g)/(a*d-b**2)
mu_nbeta = (-b*f+a*g)/(a*d-b**2)
array_mean = np.array([mu_nalpha, mu_nbeta])
print(array_mean)
print(arr2d_cov)

arr2d_pdf = multivariate_normal.pdf(x=array_ab, mean=array_mean, cov=arr2d_cov)
arr2d_pdf = arr2d_pdf.reshape(arr2d_Alpha.shape)

plt.pcolormesh(arr2d_Alpha, arr2d_Beta, arr2d_pdf, cmap='GnBu')
plt.colorbar()
```

```python
fig = plt.figure(figsize=(7, 7), dpi=120)
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(arr2d_Alpha, arr2d_Beta, arr2d_pdf, cmap='GnBu')
fig.colorbar(surf)
```

```python
from scipy.stats import norm

array_x = np.arange(-1, 1, 0.01)
array_y = np.arange(-4, 7, 0.01)
arr2d_X, arr2d_Y = np.meshgrid(array_x, array_y)
array_xy = np.c_[np.ravel(arr2d_X), np.ravel(arr2d_Y)]

list_z = []
for x,y in array_xy:
    muy = mu_nalpha * x + mu_nbeta
    sigy2 = sigma**2 + (d*(x**2)-2*b*x+a) / (a*d-b**2)
    list_z += [norm.pdf(y,muy,np.sqrt(sigy2))]

arr2d_pdf = np.array(list_z).reshape(arr2d_X.shape)

plt.pcolormesh(arr2d_X, arr2d_Y, arr2d_pdf, cmap='Reds')
plt.colorbar()
```

```python
fig = plt.figure(figsize=(7, 7), dpi=120)
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(arr2d_X, arr2d_Y, arr2d_pdf, cmap='Reds')
fig.colorbar(surf)
```
