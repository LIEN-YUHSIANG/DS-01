### Calculate contribution ratio in regression equations by python

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

array_y = np.array([20, 45, 20, 50, 45, 50])
arr2d_x = np.array([[10, 5], [10, 10], [15, 5], [15, 10],[20, 5], [20, 10]]) 

reg = LinearRegression()
reg.fit(arr2d_x, array_y)
print("Estimate beta1, 2: ", reg.coef_)
print("Estimate beta0: ", reg.intercept_)
print("Contribution ratio :", reg.score(arr2d_x, array_y))

array_predy = reg.predict(arr2d_x)
plt.scatter(array_y, array_predy)
plt.plot([20, 50], [20, 50], c='green')
```
