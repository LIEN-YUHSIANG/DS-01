### Classification problem with python

### 1. Create a plot according to flower

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
arr2d_data = np.loadtxt("irisdata.csv", delimiter=',')

array_y = arr2d_data[:, 0]
arr2d_x = arr2d_data[:, 1:]

print("Create a plot according to the flower")
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=array_y, cmap="brg")
plt.title("Plot according to flower type")
```

### 2. Fisher's linear discriminant 

```python
# import fisher's linear discriminanat 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# calculate the decision boundary
lda = LinearDiscriminantAnalysis()
lda.fit(arr2d_x, array_y)
print("Accuracy rate: ", lda.score(arr2d_x, array_y))

array_yhat = lda.predict(arr2d_x)
print(array_y)
print(array_yhat)

# Illustrate the boundary
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=array_y, cmap="brg")
x1, x2 = np.meshgrid(np.arange(0.5, 7.5, 0.02), np.arange(0, 3, 0.02))
y = lda.predict(np.stack([x1.ravel(), x2.ravel()], axis=1))
yy = y.reshape(x1.shape)
plt.pcolormesh(x1, x2, yy, alpha=0.05, cmap='brg', shading="gouraud")
plt.title("Fisher's Linear Discriminant Classification")
```

### 3. Support vector machine (SVM)

```python
# import SVC (SVM auto classification)
from sklearn.svm import SVC

svm = SVC(kernel='linear')
svm.fit(arr2d_x, array_y)
print("Accuracy rate: ", svm.score(arr2d_x, array_y))

array_yhat = svm.predict(arr2d_x)
print(array_y)
print(array_yhat)

# Illustrate the boundary
plt.scatter(arr2d_x[:, 0], arr2d_x[:, 1], c=array_y, cmap="brg")
x1, x2 = np.meshgrid(np.arange(0.5, 7.5, 0.02), np.arange(0, 3, 0.02))
y = svm.predict(np.stack([x1.ravel(), x2.ravel()], axis=1))
yy = y.reshape(x1.shape)
plt.pcolormesh(x1, x2, yy, alpha=0.05, cmap='brg', shading="gouraud")
plt.title("Fisher's Linear Discriminant Classification")

```
