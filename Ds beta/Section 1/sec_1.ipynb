```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

n, alpha, beta = 4, 2, 3
array_x = np.array([1, 4, 3, -1])
array_y = alpha * array_x + beta
print(array_x)
print(array_y)
plt.scatter(array_x, array_y)
```

```python
np.random.seed(0)
n, alpha, beta = 1000, 2, 3
sigma = 1
array_x = np.random.uniform(-1, 5, n)
array_y = alpha * array_x + beta + np.random.normal(0, sigma, n)
print(array_x)
print(array_y)
plt.scatter(array_x, array_y)


```

```python
from sklearn.linear_model import LinearRegression
arr2d_x = array_x.reshape(n, 1) # convert the array_x to 2d (nrows and 1 column)
reg = LinearRegression()
reg.fit(arr2d_x, array_y)
print(reg.coef_) # estimated alpha
print(reg.intercept_) # estimated beta
```

```python
arr2d_predX = np.array([[min(array_x)], [max(array_x)]])
array_predy = reg.predict(arr2d_predX)
print(array_predy)
plt.scatter(array_x, array_y)
plt.plot(arr2d_predX, array_predy, c='blue')
```
