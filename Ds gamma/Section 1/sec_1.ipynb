### Computing polynominal function by 2-dementaionl array

```python
import numpy as np

array_x = np.array([1, -2, 2, 0, -1])
list2d_x = []
for x in array_x:
    list2d_x += [[x, x**2, x**3]]
arr2d_x = np.array(list2d_x)
print(list2d_x)
print(arr2d_x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

beta0, beta1, beta2, beta3 = 3, -2, 0.5, 1
array_beta = np.array([beta1, beta2, beta3])
print(array_beta)
array_f = np.dot(arr2d_x, array_beta) + beta0 # 矩陣內積
print(array_f)
plt.scatter(array_x, array_f)
```

### Estimate the regression coefficient

```python
np.random.seed(1)
n = 5
array_y = array_f + np.random.normal(0, 1, n)
print(array_y)
plt.scatter(array_x, array_y, c='red')
plt.scatter(array_x, array_f)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(arr2d_x, array_y)
print(reg.coef_)
print(reg.intercept_)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
n = 100
beta0, beta1, beta2, beta3 = 3, -2, 0.5, 1
array_beta = np.array([beta1, beta2, beta3])
array_x = np.random.uniform(-2, 2, n)
list2d_x = []
for x in array_x:
    list2d_x += [[x, x**2, x**3]]
arr2d_x = np.array(list2d_x)
array_f = np.dot(arr2d_x, array_beta) + beta0
array_y = array_f + np.random.normal(0, 1, n)
plt.scatter(array_x, array_y, c='red')
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(arr2d_x, array_y)
print(reg.coef_)
print(reg.intercept_)
array_predx = np.arange(-2, 2.1, 0.1)
list2d_predx = []
for x in array_predx:
    list2d_predx += [[x, x**2, x**3]]
arr2d_predx = np.array(list2d_predx)
array_predy = reg.predict(arr2d_predx)
plt.scatter(array_x, array_y, c='red')
plt.plot(array_predx, array_predy, c='blue')
```
