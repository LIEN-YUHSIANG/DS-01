1-1 Summarizing univariate data

```python
import numpy as np
import matplotlib.pyplot as plt

# Baisc numpy
list_a = [1, 4, 6]
list_b = [3, 2, 1]
list_c = list_a + list_b

array_a = np.array(list_a) # create numerical arrays to perform calculation
array_b = np.array(list_b)
array_c = array_a + array_b

m = np.mean(array_a) # Calculate the mean

print("Addition of lists: ", list_c)
print("Addition of arrays: ", array_c)
print("Calculate the mean: ", m)
```

```python
# Load the data as array
data_array = np.loadtxt("income.csv", delimiter=",")
print(data_array)

# Calculate the mean of data
m = np.mean(data_array)
print("Mean of the data: ", m)

# Create histgram
plt.hist(data_array, bins=30) # use `bins` argument to specify the intervals
```

```python
# Find Mean, Median, Variance, and standard Deviation

m = np.mean(data_array)
med = np.median(data_array)
var = np.var(data_array)
std = np.std(data_array)

print(f"Mean: {m:.2f}")
print(f"Median: {med:.2f}")
print(f"Variance: {var:.2f}")
print(f"Standard Deviation: {std:.2f}")
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

arr2d = np.array([[-15, -5],  # row 0
                  [23, 45],  # row 1
                  [-13, -23],
                  [-76, -67]])

array_x = arr2d[:, 0]
array_y = arr2d[:, 1]
print("arr2d: \n", arr2d)
print("array_x: ", array_x)
print("array_y: ", array_y)

plt.scatter(array_x, array_y, s=10, c='blue', alpha=0.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt

arr2d = np.loadtxt("2d.csv", delimiter=",")

array_x = arr2d[:, 0]
array_y = arr2d[:, 1]

# Create a scatter plot
plt.scatter(array_x, array_y, s=10, c='blue', alpha=0.5)

```

```python
# Create a 2d histogram
H = plt.hist2d(array_x, array_y, cmap='Blues')
plt.colorbar(H[3]) # Express intercal frequencies
```

```python
# Caluculate the correlation coefficient
arr2d_cov = np.cov(array_x, array_y, bias=True)
arr2d_r = np.corrcoef(array_x, array_y)
print("Covariance:\n", arr2d_cov)
print("Correlation Coefficient: \n", arr2d_r)
```

1-4 Practice exercise (Bivariate data)

```python
import numpy as np

arr2d = np.array([[-15, -5], [23, 45], [-13, -23] ,[-76, -67]])
array_x = arr2d[:, 0]
array_y = arr2d[:, 1]

arr2d_cov = np.cov(array_x, array_y, bias=True)
arr2d_r = np.corrcoef(array_x, array_y)

cov_xy = arr2d_cov[0, 1]
r_xy = arr2d_r[0, 1]

print((' {0:.3f}').format(cov_xy))
print((' {0:.3f}').format(r_xy))

```
