```python
# Draw graphs and the characteristics of a normal distribution
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

array_x = np.arange(-3, 4)
array_f = -array_x * array_x
print('array_x= ', array_x)
print("array_f= ", array_f)
plt.plot(array_x, array_f) # arrya_x = Horizontal axis, array_f = Vertocal axis
```

```python
array_x = np.arange(-3, 3.1, 0.2) # Make the graph smoother by making the interval smoother
array_f = -array_x * array_x
plt.plot(array_x, array_f)
```

```python
mu = 4 # push the graph to the right
array_x = np.arange(-3+mu, 3.1+mu, 0.2) 
array_f = -(array_x-mu) * (array_x - mu)
plt.plot(array_x, array_f)
```

```python
# Show 2 graphs at the same time

mu = 4 # push the graph to the right
array_x = np.arange(-3, 3.1+mu, 0.2) 
array_f1 = -array_x * array_x
array_f2 = -(array_x - mu) * (array_x - mu)

plt.plot(array_x, array_f1)
plt.plot(array_x, array_f2)
```

```python
# Enlarge graph horizontally by sigma time

sigma = 2
array_x = np.arange(-3*sigma, 3.01*sigma, 0.2)
array_f = -(array_x/sigma) * (array_x/sigma)
plt.plot(array_x, array_f)
```

```python
# Enlarging graph horizontally by sigma times, and carry out parallel transition by mu

mu, sigma = 4, 2
array_x = np.arange(-3*sigma+mu, 3.01*sigma+mu, 0.2)
array_f = -(array_x-mu)/sigma * (array_x-mu)/sigma
plt.plot(array_x, array_f)
```

### Draw graph for normal distribution

```python
# Draw graph for normal distribution

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

array_x = np.arange(-3, 3.01, 0.1)
array_f = np.e**(-array_x * array_x)
plt.plot(array_x, array_f)
```

```python
# lower the height of graph by 1/sqrt(2pi) and expand horizontally by sqrt(2) times.

array_x = np.arange(-3, 3.01, 0.1)
array_f = np.e**(-array_x * array_x)
plt.plot(array_x, array_f)

array_x2 = np.arange(-4, 4.01, 0.1)
array_f2 = 1/np.sqrt(2**np.pi) * (np.e**(-array_x2 * array_x2/2))
plt.plot(array_x2, array_f2)
```

```python
array_x1 = np.arange(-4, 4.01, 0.1)
array_f1 = np.e**(-array_x1 * array_x1)
plt.plot(array_x1, array_f1)

mu, sigma = 4, 2
array_x = np.arange(-4*sigma+mu, 4.01*sigma+mu, 0.1)
array_f = 1/np.sqrt(2*np.pi*sigma*sigma) * (np.e**(-1/2 * (array_x - mu)/sigma * (array_x - mu)/sigma))
plt.plot(array_x, array_f)
```
