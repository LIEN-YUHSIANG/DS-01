3-1 Generate Pseudo-random number sequence

```python
# Generate random numbers in Bernoulli Distribution

%matplotlib inline
from re import X
import numpy as np 
import matplotlib.pyplot as plt

np.random.seed(0)
size = 50
theta = 0.1 
array_x = np.random.binomial(1, theta, size)
print(array_x)
cnt = sum(array_x)
array_hor = np.array([0, 1])
array_ver = np.array([size-cnt, cnt])
plt.bar(array_hor, array_ver, tick_label=array_hor)
```

```python
# Generate random numbers in a descrete uniform distribution

np.random.seed(0)
size  = 50
a = 1
m = 6
array_x = np.random.randint(a, a+m, size)
print(array_x)
array_hor = np.array(range(a, a+m))
array_ver, array_edges = np.histogram(array_x, bins=range(a, a+m+1))
plt.bar(array_hor, array_ver, tick_label=array_hor)
```

```python
# Generate random numbers in a Poisson distribution

np.random.seed(0)
size =100
lam = 5
array_x = np.random.poisson(lam, size)
print(array_x)
plt.hist(array_x, bins=range(0, 15))
```

3-2 Generate random numbers for continuous probability distributions.

```python
# Generate random number with Uniform distribution

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
size = 50
alpha, beta = 3.0, 6.0
array_x = np.random.uniform(alpha, beta, size)

np.set_printoptions(precision=1)
print(array_x)
plt.hist(array_x)

```

```python
# Generate random number with Exponential distribution

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
size = 100
lam = 0.2
array_x = np.random.exponential(1/lam, size)

np.set_printoptions(precision=1)
print(array_x)
plt.hist(array_x)
```

```python
# Generate random number with Normal distribution

%matplotlib inline
from xml.etree.ElementPath import prepare_self
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
size = 100
mu, sig = 4, 2
array_x = np.random.normal(mu, sig, size)

np.set_printoptions(precision=1)
print(array_x)
plt.hist(array_x, bins=range(-4, 13))
```
